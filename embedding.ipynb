```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
###sentences
sent = ['the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'I am a good boy',
        'I am a good developer',
        'understand the meaning of words',
        'your videos are good',]
```

```python
sent
```

```python
## define the vocabulary size
voc_size = 10000
```

```python
##One hot representation(optional)

onehot_repr = [one_hot(words,voc_size) for words in sent]
print(onehot_repr)
```

```python
#word embedding representation
from tensorflow.keras.layers import Embedding 
#from tensorflow.keras.processing import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential#first we made the sequential model
import numpy as np

```

```python
# as we need to mke all the sentences of same length so we will use pad_sequences(to train)
sent_length = 8
embedded_docs = pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)#padding='pre' means padding will be done before the sentence means zeroes will be added before the sentence to make it of same length
print(embedded_docs)

```

```python
##feature representation
dim = 10#dimension of the feature

# creating the model
model = Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')

```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
